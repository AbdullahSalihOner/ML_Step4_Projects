```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('kanser.csv') #read data from csv file    / veri okuma
```

```python
data.head() #show first 5 rows of data / ilk 5 satırı göster
```

```python
data.drop(['id','Unnamed: 32'],axis=1,inplace=True) #drop id and Unnamed: 32 columns / id ve Unnamed: 32 sütunlarını sil
data.head() #show first 5 rows of data / ilk 5 satırı göster
```

```python
M = data[data.diagnosis == 'M'] #Malignant / Kötü huylu
B = data[data.diagnosis == 'B'] #Benign / İyi huylu
```

```python
plt.scatter(M.radius_mean,M.texture_mean,color="red",label="Malignant",alpha=0.4)
plt.scatter(B.radius_mean,B.texture_mean,color="green",label="Benign",alpha=0.4)    
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
plt.legend()
plt.show()
```

```python
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis] #M=1, B=0
y = data.diagnosis.values   
x = data.drop(["diagnosis"],axis=1) #drop diagnosis column   / diagnosis sütununu sil
data.head() 
```

```python
from sklearn.model_selection import train_test_split #split data for train and test / veriyi eğitim ve test için ayırma
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.15,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler #standardization / standartlaştırma
sc = StandardScaler()
```

```python
xtrain = sc.fit_transform(x_train) #fit and transform train data / eğitim verisini fit ve transform et
xtest = sc.transform(x_test) #transform test data / test verisini transform et
```

```python
from sklearn.tree import DecisionTreeClassifier #decision tree classifier    / decision tree sınıflandırıcı
dc = DecisionTreeClassifier(max_depth=5,random_state=42)
```

```python
dc.fit(xtrain,y_train) #fit model / modeli eğit
```

```python
yhead = dc.predict(xtest) #predict test data / test verisini tahmin et
```

```python
dc.score(xtest,y_test) #accuracy score / doğruluk skoru
```

```python
from sklearn.metrics import confusion_matrix #confusion matrix / karışıklık matrisi
cm = confusion_matrix(y_test,yhead)   
print(cm)
```

```python
import seaborn as sns #confusion matrix visualization    / karışıklık matrisi görselleştirme
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(cm,annot=True,linewidths=0.5,linecolor="red",fmt=".0f",ax=ax)   
plt.xlabel("y_head") #y_head = prediction / tahmin
plt.ylabel("y_true") #y_true = actual / gerçek
plt.show()
```
